Data Extraction

```python
X=data.drop("Target",axis=1)
X
```

```python
Y=data['Target']
Y
```

Data Splitting

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X,Y ,test_size=0.3,random_state=42)
# Random state for prevent shuffling
# Test size 0.3 means 30% for test and 70% for train
```

Model Creation

```python
data.dtypes
```

```python
cat_features=X.select_dtypes("object").columns.tolist()
print("Totl Categorical Features: ",len(data.select_dtypes("object").columns.tolist()))
cat_features
```

```python
from sklearn.preprocessing import LabelEncoder
encoder=LabelEncoder()
for entry in cat_features:
    data[entry]=encoder.fit_transform(data[entry])
    print(data[entry].unique())
```

```python
from sklearn.preprocessing import LabelEncoder
encoder=LabelEncoder()
for entry in cat_features:
    X_train[entry]=encoder.fit_transform(X_train[entry])
    X_test[entry]=encoder.transform(X_test[entry])
    print(X_train[entry].unique())
```

```python
pd.DataFrame(X_train)
```

```python
from sklearn.linear_model import LogisticRegression
clf=LogisticRegression()
```

```python
clf.fit(X_train, Y_train)
```

Prediction for Random Input

```python
Y_pred=clf.predict(X_test)
```

```python
clf.predict([[4,3,4,2,24,4,2,5,2,4,2,4,2,24]])
```

Model Performance Evaluation

```python
print("Accuracy: ",clf.score(X_train,Y_train))
print("Accuracy: ",clf.score(X_test,Y_test))
```

```python
from sklearn.metrics import root_mean_squared_error
root_mean_squared_error(Y_test,Y_pred)
```
