```python
import pandas as pd
data=pd.read_csv('Dataset\Mall_Customers.csv')
data.head()
```

```python
data.columns
```

```python
data.isnull().sum()
```

```python
data_in = data[['Annual Income (k$)','Spending Score (1-100)']]
```

```python
import seaborn as sns
sns.scatterplot(x="Annual Income (k$)",y="Spending Score (1-100)",data=data_in)
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=5, random_state=0)
kmeans.fit(data_in)
labels=kmeans.predict(data_in)
data_in['Cluster']=labels
data_in.head()
```

```python
sns.scatterplot(x="Annual Income (k$)",y="Spending Score (1-100)",data=data_in,
                hue='Cluster',palette=['red', 'green','blue','yellow','orange'])
```
