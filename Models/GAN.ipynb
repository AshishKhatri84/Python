## **GAN**

```python
import tensorflow as tf
from tensorflow.keras import layers
import numpy as np
import matplotlib.pyplot as plt

# Load and preprocess CIFAR-10 dataset
(x_train, _), (_, _) = tf.keras.datasets.cifar10.load_data()
x_train = (x_train.astype('float32') - 127.5) / 127.5  # Normalize to [-1, 1]
x_train = np.reshape(x_train, (x_train.shape[0], 32, 32, 3))

# Hyperparameters
latent_dim = 100
batch_size = 128
epochs = 100  # Reduced from 10,000 to 100
display_interval = 10  # Show images every 10 epochs

# Build Generator
def build_generator():
    model = tf.keras.Sequential([
        layers.Dense(256 * 4 * 4, input_dim=latent_dim, kernel_initializer="he_normal"),
        layers.ReLU(),
        layers.Reshape((4, 4, 256)),
        layers.Conv2DTranspose(128, kernel_size=4, strides=2, padding='same', kernel_initializer="he_normal"),
        layers.BatchNormalization(),
        layers.ReLU(),
        layers.Conv2DTranspose(64, kernel_size=4, strides=2, padding='same', kernel_initializer="he_normal"),
        layers.BatchNormalization(),
        layers.ReLU(),
        layers.Conv2DTranspose(3, kernel_size=4, strides=2, padding='same', activation='tanh', kernel_initializer="he_normal")
    ])
    return model

# Build Discriminator
def build_discriminator():
    model = tf.keras.Sequential([
        layers.Conv2D(64, kernel_size=3, strides=2, padding='same', input_shape=(32, 32, 3), kernel_initializer="he_normal"),
        layers.LeakyReLU(alpha=0.2),
        layers.Conv2D(128, kernel_size=3, strides=2, padding='same', kernel_initializer="he_normal"),
        layers.BatchNormalization(),
        layers.LeakyReLU(alpha=0.2),
        layers.Conv2D(256, kernel_size=3, strides=2, padding='same', kernel_initializer="he_normal"),
        layers.BatchNormalization(),
        layers.LeakyReLU(alpha=0.2),
        layers.Flatten(),
        layers.Dense(1, activation='sigmoid')
    ])
    return model

# Initialize models
generator = build_generator()
discriminator = build_discriminator()

# Compile Discriminator
discriminator.compile(optimizer=tf.keras.optimizers.Adam(0.0002, 0.5), loss='binary_crossentropy', metrics=['accuracy'])

# Compile GAN (Generator + Discriminator)
discriminator.trainable = False
z = layers.Input(shape=(latent_dim,))
img = generator(z)
valid = discriminator(img)
gan = tf.keras.Model(z, valid)
gan.compile(optimizer=tf.keras.optimizers.Adam(0.0002, 0.5), loss='binary_crossentropy')

# Function to display generated images
def display_images(epoch):
    noise = np.random.normal(0, 1, (16, latent_dim))
    gen_imgs = generator.predict(noise)
    gen_imgs = 0.5 * gen_imgs + 0.5  # Rescale to [0,1]

    fig, axs = plt.subplots(4, 4, figsize=(6, 6))
    for i in range(4):
        for j in range(4):
            axs[i, j].imshow(gen_imgs[i * 4 + j])
            axs[i, j].axis('off')

    plt.suptitle(f"Generated Images at Epoch {epoch}")
    plt.show()

# Training Loop
losses = {"d_loss": [], "g_loss": []}

for epoch in range(epochs):
    idx = np.random.randint(0, x_train.shape[0], batch_size)
    real_imgs = x_train[idx]

    # Generate fake images
    noise = np.random.normal(0, 1, (batch_size, latent_dim))
    fake_imgs = generator.predict(noise)

    # Labels
    real_labels = np.ones((batch_size, 1)) * 0.9  # Label smoothing
    fake_labels = np.zeros((batch_size, 1))

    # Train Discriminator
    d_loss_real = discriminator.train_on_batch(real_imgs, real_labels)
    d_loss_fake = discriminator.train_on_batch(fake_imgs, fake_labels)
    d_loss = 0.5 * np.add(d_loss_real, d_loss_fake)

    # Train Generator (Trick Discriminator)
    noise = np.random.normal(0, 1, (batch_size, latent_dim))
    g_loss = gan.train_on_batch(noise, np.ones((batch_size, 1)))

    # Store Losses
    losses["d_loss"].append(d_loss[0])
    losses["g_loss"].append(g_loss)

    # Print Progress
    if epoch % 10 == 0:
        print(f"Epoch {epoch}/{epochs} | D Loss: {d_loss[0]:.4f} | G Loss: {g_loss:.4f}")

    # Display Sample Images
    if epoch % display_interval == 0:
        display_images(epoch)

# Plot Loss Curves
plt.figure(figsize=(10, 5))
plt.plot(losses["d_loss"], label="Discriminator Loss", color="red")
plt.plot(losses["g_loss"], label="Generator Loss", color="blue")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.title("GAN Training Loss")
plt.legend()
plt.grid(True)
plt.show()
```

## **SRGAN**

```python
import tensorflow as tf
from tensorflow.keras import layers, Model
import numpy as np
import matplotlib.pyplot as plt
import cv2

# Load CIFAR-10 dataset (low-res images)
(x_train, _), (_, _) = tf.keras.datasets.cifar10.load_data()
x_train = x_train.astype("float32") / 255.0  # Normalize to [0,1]

# Create low-resolution images by downscaling
def downscale_images(images, scale=2):
    lr_images = np.array([cv2.resize(img, (img.shape[1] // scale, img.shape[0] // scale)) for img in images])
    return lr_images

lr_train = downscale_images(x_train, scale=2)  # Convert to 16x16 images
lr_train = np.reshape(lr_train, (-1, 16, 16, 3))

# Hyperparameters
epochs = 100
batch_size = 128
display_interval = 10  # Show results every 10 epochs
lr_shape = (16, 16, 3)
hr_shape = (32, 32, 3)
latent_dim = 100

# **Build Generator (SRGAN)**
def build_generator():
    inputs = layers.Input(shape=lr_shape)

    # Initial Convolution
    x = layers.Conv2D(64, kernel_size=3, padding="same")(inputs)
    x = layers.PReLU(shared_axes=[1, 2])(x)

    # Residual Blocks
    def residual_block(x):
        res = layers.Conv2D(64, kernel_size=3, padding="same")(x)
        res = layers.BatchNormalization()(res)
        res = layers.PReLU(shared_axes=[1, 2])(res)
        res = layers.Conv2D(64, kernel_size=3, padding="same")(res)
        res = layers.BatchNormalization()(res)
        return layers.Add()([x, res])

    for _ in range(5):
        x = residual_block(x)

    # Upsampling Blocks
    x = layers.Conv2D(256, kernel_size=3, padding="same")(x)
    x = layers.UpSampling2D(size=2)(x)
    x = layers.PReLU(shared_axes=[1, 2])(x)

    # Output Layer
    outputs = layers.Conv2D(3, kernel_size=3, padding="same", activation="tanh")(x)
    return Model(inputs, outputs, name="Generator")

# **Build Discriminator**
def build_discriminator():
    inputs = layers.Input(shape=hr_shape)
    x = layers.Conv2D(64, kernel_size=3, strides=2, padding="same")(inputs)
    x = layers.LeakyReLU(alpha=0.2)(x)

    x = layers.Conv2D(128, kernel_size=3, strides=2, padding="same")(x)
    x = layers.BatchNormalization()(x)
    x = layers.LeakyReLU(alpha=0.2)(x)

    x = layers.Conv2D(256, kernel_size=3, strides=2, padding="same")(x)
    x = layers.BatchNormalization()(x)
    x = layers.LeakyReLU(alpha=0.2)(x)

    x = layers.Flatten()(x)
    x = layers.Dense(1, activation="sigmoid")(x)
    return Model(inputs, x, name="Discriminator")

# Initialize Models
generator = build_generator()
discriminator = build_discriminator()

# Compile Discriminator
discriminator.compile(optimizer=tf.keras.optimizers.Adam(0.0002, 0.5), loss="binary_crossentropy", metrics=["accuracy"])

# Create and Compile GAN (SRGAN)
discriminator.trainable = False
input_lr = layers.Input(shape=lr_shape)
generated_hr = generator(input_lr)
validity = discriminator(generated_hr)
gan = Model(input_lr, validity)
gan.compile(optimizer=tf.keras.optimizers.Adam(0.0002, 0.5), loss="binary_crossentropy")

# **Function to Display Super-Resolved Images**
def display_images(epoch):
    idx = np.random.randint(0, lr_train.shape[0], 4)
    low_res_imgs = lr_train[idx]
    high_res_fake = generator.predict(low_res_imgs)

    low_res_imgs = np.clip(low_res_imgs, 0, 1)  # Normalize LR images
    high_res_fake = (high_res_fake + 1) / 2  # Rescale from [-1,1] to [0,1]

    fig, axs = plt.subplots(2, 4, figsize=(8, 4))
    for i in range(4):
        axs[0, i].imshow(low_res_imgs[i])
        axs[0, i].axis("off")
        axs[1, i].imshow(high_res_fake[i])
        axs[1, i].axis("off")

    axs[0, 0].set_title("Low-Res Input")
    axs[1, 0].set_title("Super-Resolved Output")
    plt.suptitle(f"Super-Resolved Images at Epoch {epoch}")
    plt.show()

# **Training Loop**
losses = {"d_loss": [], "g_loss": []}

for epoch in range(epochs):
    idx = np.random.randint(0, lr_train.shape[0], batch_size)
    lr_imgs = lr_train[idx]
    hr_imgs = x_train[idx]

    # Generate fake high-resolution images
    fake_hr_imgs = generator.predict(lr_imgs)

    # Labels
    real_labels = np.ones((batch_size, 1)) * 0.9  # Label smoothing
    fake_labels = np.zeros((batch_size, 1))

    # Train Discriminator
    d_loss_real = discriminator.train_on_batch(hr_imgs, real_labels)
    d_loss_fake = discriminator.train_on_batch(fake_hr_imgs, fake_labels)
    d_loss = 0.5 * np.add(d_loss_real, d_loss_fake)

    # Train Generator
    g_loss = gan.train_on_batch(lr_imgs, np.ones((batch_size, 1)))

    # Store Losses
    losses["d_loss"].append(d_loss[0])
    losses["g_loss"].append(g_loss)

    # Print Progress
    if epoch % 10 == 0:
        print(f"Epoch {epoch}/{epochs} | D Loss: {d_loss[0]:.4f} | G Loss: {g_loss:.4f}")

    # Display Images
    if epoch % display_interval == 0:
        display_images(epoch)

# **Plot Loss Graph**
plt.figure(figsize=(10, 5))
plt.plot(losses["d_loss"], label="Discriminator Loss", color="red")
plt.plot(losses["g_loss"], label="Generator Loss", color="blue")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.title("SRGAN Training Loss")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import tensorflow as tf
from tensorflow.keras import layers
import numpy as np
import matplotlib.pyplot as plt

# Load and preprocess CIFAR-10 dataset (Low-res and High-res)
(x_train, _), (_, _) = tf.keras.datasets.cifar10.load_data()
x_train = (x_train.astype('float32') - 127.5) / 127.5  # Normalize to [-1, 1]

# Create low-res images by downscaling
lr_train = tf.image.resize(x_train, (16, 16))  # Low-resolution (16x16)
lr_train = tf.image.resize(lr_train, (32, 32))  # Upscale back to match input size

# Hyperparameters
batch_size = 64
epochs = 1000
sample_interval = 100  # Display images every 100 epochs

# Build SRGAN Generator
def build_generator():
    model = tf.keras.Sequential([
        layers.Conv2D(64, kernel_size=3, strides=1, padding='same', input_shape=(32, 32, 3)),
        layers.PReLU(),
        layers.Conv2D(128, kernel_size=3, strides=1, padding='same'),
        layers.BatchNormalization(),
        layers.PReLU(),
        layers.Conv2D(256, kernel_size=3, strides=1, padding='same'),
        layers.BatchNormalization(),
        layers.PReLU(),
        layers.Conv2D(3, kernel_size=3, strides=1, padding='same', activation='tanh')
    ])
    return model

# Build Discriminator
def build_discriminator():
    model = tf.keras.Sequential([
        layers.Conv2D(64, kernel_size=3, strides=2, padding='same', input_shape=(32, 32, 3)),
        layers.LeakyReLU(alpha=0.2),
        layers.Conv2D(128, kernel_size=3, strides=2, padding='same'),
        layers.BatchNormalization(),
        layers.LeakyReLU(alpha=0.2),
        layers.Conv2D(256, kernel_size=3, strides=2, padding='same'),
        layers.BatchNormalization(),
        layers.LeakyReLU(alpha=0.2),
        layers.Flatten(),
        layers.Dense(1, activation='sigmoid')
    ])
    return model

# Initialize models
generator = build_generator()
discriminator = build_discriminator()

# Compile Discriminator
discriminator.compile(optimizer=tf.keras.optimizers.Adam(0.0002, 0.5), loss='binary_crossentropy', metrics=['accuracy'])

# Compile GAN (Generator + Discriminator)
discriminator.trainable = False
lr_input = layers.Input(shape=(32, 32, 3))
sr_output = generator(lr_input)
validity = discriminator(sr_output)
gan = tf.keras.Model(lr_input, validity)
gan.compile(optimizer=tf.keras.optimizers.Adam(0.0002, 0.5), loss='binary_crossentropy')

# Display Generated Images
def display_images(epoch):
    idx = tf.random.uniform(shape=(4,), minval=0, maxval=lr_train.shape[0], dtype=tf.int32)  # Generate indices as TensorFlow tensor
    lr_imgs = tf.gather(lr_train, idx)  # Use tf.gather to select elements
    hr_imgs = tf.gather(x_train, idx)  # Use tf.gather for hr_imgs as well
    sr_imgs = generator.predict(lr_imgs)

    lr_imgs = (lr_imgs + 1) / 2  # Convert back to [0,1]
    hr_imgs = (hr_imgs + 1) / 2
    sr_imgs = (sr_imgs + 1) / 2

    fig, axs = plt.subplots(3, 4, figsize=(8, 6))
    fig.suptitle(f"Super-Resolved Images at Epoch {epoch}", fontsize=14)

    for i in range(4):
        axs[0, i].imshow(lr_imgs[i])
        axs[0, i].axis("off")
        axs[0, i].set_title("Low-Res Input")

        axs[1, i].imshow(sr_imgs[i])
        axs[1, i].axis("off")
        axs[1, i].set_title("Super-Resolved Output")

        axs[2, i].imshow(hr_imgs[i])
        axs[2, i].axis("off")
        axs[2, i].set_title("High-Res Ground Truth")

    plt.show()

# Training Loop
losses = {"d_loss": [], "g_loss": []}

for epoch in range(epochs):
    idx = tf.random.uniform(shape=(batch_size,), minval=0, maxval=lr_train.shape[0], dtype=tf.int32)  # Generate random indices as a TensorFlow tensor
    lr_imgs = tf.gather(lr_train, idx)  # Use tf.gather to select elements from lr_train based on idx
    hr_imgs = tf.gather(x_train, idx)  # Use tf.gather for hr_imgs as well

    # Generate fake high-resolution images
    sr_imgs = generator.predict(lr_imgs)

    # Labels
    real_labels = np.ones((batch_size, 1)) * 0.9  # Label smoothing
    fake_labels = np.zeros((batch_size, 1))

    # Train Discriminator
    d_loss_real = discriminator.train_on_batch(hr_imgs, real_labels)
    d_loss_fake = discriminator.train_on_batch(sr_imgs, fake_labels)
    d_loss = 0.5 * np.add(d_loss_real, d_loss_fake)

    # Train Generator
    g_loss = gan.train_on_batch(lr_imgs, np.ones((batch_size, 1)))

    # Store Losses
    losses["d_loss"].append(d_loss[0])
    losses["g_loss"].append(g_loss)

    # Print Progress
    if epoch % 100 == 0:
        print(f"Epoch {epoch}/{epochs} | D Loss: {d_loss[0]:.4f} | G Loss: {g_loss:.4f}")
        display_images(epoch)  # Display images every 100 epochs

# Plot Loss Curves
plt.figure(figsize=(10, 5))
plt.plot(losses["d_loss"], label="Discriminator Loss", color="red")
plt.plot(losses["g_loss"], label="Generator Loss", color="blue")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.title("SRGAN Training Loss")
plt.legend()
plt.grid(True)
plt.show()
```
