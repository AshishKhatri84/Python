Multi-variable Linear Regression

```python
import pandas as pd
data2=pd.read_csv("Dataset\\advertising.csv")
data2
```

Data Visualization

```python
import matplotlib.pyplot as plt
plt.scatter(data2["TV"],data2["Sales"])
plt.scatter(data2["Radio"],data2["Sales"])
plt.scatter(data2["Newspaper"],data2["Sales"])
```

```python
import seaborn as sns
sns.scatterplot(x='TV',y='Sales',data=data2)
sns.scatterplot(x='Radio',y='Sales',data=data2)
sns.scatterplot(x='Newspaper',y='Sales',data=data2)
```

Data Extraction

```python
X=data2.drop({'Sales'},axis=1)
Y=data2['Sales']
print(X)
print(Y)
```

Data Splitting

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X,Y ,test_size=0.3,random_state=42)
```

Model Creation

```python
from sklearn.linear_model import LinearRegression
clf=LinearRegression()
```

```python
clf.fit(X_train,Y_train)
```

Prediction for Random Input

```python
Y_pred=clf.predict(X_test)
```

```python
clf.predict([[10,20,30]])
```

Model Performance Evaluation

```python
print("Accuracy: ",clf.score(X_train,Y_train))
print("Accuracy: ",clf.score(X_test,Y_test))
```

```python
from sklearn.metrics import root_mean_squared_error
root_mean_squared_error(Y_test,Y_pred)
```
